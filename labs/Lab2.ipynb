### Бинарный поиск

```python
# O(log (n))

import time
import random

start_time = time.time()


def create_array():
    return [random.randint(0, 100) for value in range(random.randint(5, 15))]


def binary_search(array, item):
    array.sort()
    low = 0
    high = len(array) - 1

    while low <= high:  # Пока эта часть не сократится до одного элемента, проверяем средний элемент
        mid = (low + high) // 2
        guess = array[mid]

        if guess == item:  # Значение найдено
            return mid
        elif guess > item:  # Много
            high = mid - 1
        else:  # Мало
            low = mid + 1
    return None  # Значение не существует


def binary_search_pop(array, item):
    index = binary_search(array, item)
    if index:
        array.pop(index)
    else:
        print('Данного элемента нет в массиве')
    return array


def binary_search_append(array, item):
    index = binary_search(array, item)
    if index != None and array[index] == item:
        print('Данный элемент уже существует в массиве')
    else:
        i = 0
        while i < len(array):
            if item < array[0]:  # Меньше первого
                array.insert(0, item)

            if item > array[len(array) - 1]:  # Больше последнего
                array.insert(len(array), item)

            if item > array[i] and item < array[i + 1]:  # Где-то по середине
                array.insert(i + 1, item)

            i += 1
    return array


array = create_array()
print('Исходный массив:', array)

array.sort()
print('Отсортированный массив:', array)

search_value = binary_search(array, int(input('Введите значение для поиска элемента: ')))
print('Индекс запрошенного элемента: ', search_value)

pop_value = binary_search_pop(array, int(input('Введите значение для удаления элемента: ')))
print('Исходный массив:', pop_value)

append_value = binary_search_append(array, int(input('Введите значение для добавления элемента: ')))
print('Исходный массив:', append_value)

print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))     # Убрать input() для проверки времени
```

### Бинарное дерево

```python
# O(log (n)) - mid
# O(n) - worst

class TreeNode:
    
    def __init__(self, data):
        """ Инициализация узла """
        self.data = data
        self.left = None
        self.right = None

    
    def insert(self, data):
        """ Создание вершин дерева. Сравнивает новое значение с родительским узлом и решает,
            добавить ли его в дерево как левый узел или как правый """
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = TreeNode(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = TreeNode(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data


    def preorder_traversal(self, root):
        """ Прямой обход бинарного дерева
            Root -> Left -> Right """
        arr = []
        if root:
            arr.append(root.data)
            arr = arr + self.preorder_traversal(root.left)
            arr = arr + self.preorder_traversal(root.right)
        return arr

    
    def search(self, data):
        """ Нахождения вершины с запрошенным значением """
        if self.data:
            # Если цель совпадает со значением узла, то элемент найден
            if data == self.data:
                return (f'Элемент со значением {data} найден')
            elif data < self.data:
                # Если ветка пустая, то элемента в ней нет, иначе дальше происходит поиск элемента
                if self.left is None:
                    return (f'Элемент со значением {data} не найден')
                return self.left.search(data)
            else:
                if self.right is None:
                    return (f'Элемент со значением {data} не найден')
                return self.right.search(data)
        else:
            return None

    
    def delete(self, data):
        if self.data:
            # Если ключ больше/ меньше значения узла
            if data < self.data:
                # Вызываем функцию для нового узла
                self.left = self.left.delete(data)
            elif data > self.data:
                self.right = self.right.delete(data)
            # Если у узла есть обе ветки, помещаем в узел наименьший элемент из правой ветки и удаляем его из ветки
            elif (self.left is not None) and (self.right is not None):
                self.data = min(self.right).data
                self.right = self.right.delete(self.data)
            else:
                # Если есть только одна ветка, то элемент удаляется, при этом узел левой ветки становится на его место
                if self.left is not None:
                    self = self.left
                elif self.right is not None:
                    self = self.right
                else:
                    self = None
        return self

    
    def print_tree(self):
        """ Просто функция, выводящая дерево построчно """
        if self.left:
            self.left.print_tree()
        print(self.data)
        if self.right:
            self.right.print_tree()
            
    
    def display(self):
        """ Визуальное представление дерева. Код взят с сайта:
        https://stackoverflow.com/questions/34012886/print-binary-tree-level-by-level-in-python """
        lines, *_ = self._display_aux()
        for line in lines:
            print(line)

            
    def _display_aux(self):
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if self.right is None and self.left is None:
            line = '%s' % self.data
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle
        # Only left child.
        if self.right is None:
            lines, n, p, x = self.left._display_aux()
            s = '%s' % self.data
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2
        # Only right child.
        if self.left is None:
            lines, n, p, x = self.right._display_aux()
            s = '%s' % self.data
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2
        # Two children.
        left, n, p, x = self.left._display_aux()
        right, m, q, y = self.right._display_aux()
        s = '%s' % self.data
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2


root = TreeNode(60)
root.insert(35)
root.insert(21)
root.insert(76)
root.insert(89)
root.insert(71)
root.insert(24)
root.insert(68)
root.insert(17)
root.insert(23)
root.insert(42)
root.insert(63)
root.insert(11)
root.insert(69)

root.display()
print()

print(root.preorder_traversal(root),'\n')
print(root.search(21))

root.delete(71)
print()

root.display()
print()

print(root.preorder_traversal(root))
# root.print_tree()
```

### Метод Фибоначчи

```python
# O(log (n))

import copy
import time

start_time = time.time()


def fibonacci_search(arr, element):
    arr.sort()
    
    offset = -1     # Смещение
    fib1 = 0
    fib2 = 1
    fib_sum = fib1 + fib2
    
    while fib_sum < len(arr):
        fib1 = fib2
        fib2 = fib_sum
        fib_sum = fib1 + fib2
        
    while fib_sum > 1:
        index = min(offset + fib1, len(arr) - 1)
        if element > arr[index]:
            fib_sum = fib2
            fib2 = fib1
            fib1 = fib_sum - fib2
            offset = index
        elif element < arr[index]:
            fib_sum = fib1
            fib2 = fib2 - fib1
            fib1 = fib_sum - fib2
        else:
            return index
        
    if fib2 and arr[offset + 1] == element:
        return offset
    
    return None


def fibonacci_delete(arr, element):
    arr.sort()
    index = fibonacci_search(arr, element)
    if index is not None:
        arr.pop(index)
    return arr


def fibonacci_insert(arr, element):
    """ При помощи метода Фибоначчи определяем место вставки нашего элемента в отсортированном списке """
    arr.sort()
    
    offset = -1
    fib1 = 0
    fib2 = 1
    fib_sum = fib1 + fib2

    while fib_sum < len(arr):
        fib1 = fib2
        fib2 = fib_sum
        fib_sum = fib1 + fib2
        
    while fib_sum > 1:
        index = min(offset + fib1, len(arr) - 1)
        if element > arr[index]:
            fib_sum = fib2
            fib2 = fib1
            fib1 = fib_sum - fib2
            offset = index
        elif element < arr[index]:
            fib_sum = fib1
            fib2 = fib2 - fib1
            fib1 = fib_sum - fib2
        else:
            arr.insert(index, element)
            return index
        
    if (offset + 1 == len(arr) - 1) and (arr[offset + 1] < element):
        offset += 1
        
    arr.insert(offset + 1, element)
    return offset + 1


fibonacci_array = copy.deepcopy(array)

print('Исходный массив:', fibonacci_array)
print('Индекс запрошенного элемента:', fibonacci_search(fibonacci_array, 88))
print('Исходный массив:', fibonacci_delete(fibonacci_array, 23))
print('Индекс добавленного элемента:', fibonacci_insert(fibonacci_array, 55))
print('Исходный массив:', fibonacci_array)

print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
```

### Интерполяционный метод

```python
# O(log (n)) - mid
# O(n) - worst

import copy
import time

start_time = time.time()


def interpolation_search(arr, element):
    arr.sort()
    left = 0                # Индекс первого элемента последовательности
    right = len(arr) - 1    # Индекс последнего элемента последовательности
    
    while (arr[left] < element) and (arr[right] > element):
        if arr[left] == arr[right]:
            break
        index = (((element - arr[left]) * (left - right)) // (arr[left] - arr[right])) + left
        if arr[index] > element:
            right = index - 1
        elif arr[index] < element:
            left = index + 1
        else:
            return index

    if arr[left] == element:
        return left
    if arr[right] == element:
        return right
    return None


def interpolation_delete(arr, element):
    index = interpolation_search(arr, element)
    if index:
        arr.pop(index)
    return arr


def interpolation_insert(arr, element):
    arr.sort()
    left = 0
    right = len(arr) - 1
    while (arr[left] < element) and (arr[right] > element):
        if arr[left] == arr[right]:
            break
        index = (((element - arr[left]) * (left - right)) // (arr[left] - arr[right])) + left
        if arr[index] > element:
            right = index - 1
            continue
        elif arr[index] < element:
            left = index + 1
        else:
            arr.insert(index, element)
            return arr
            
    if arr[left] > element:
        arr.insert(left, element)
    elif arr[right] < element:
        arr.insert(right + 1, element)
    return arr

interpolation_array = copy.deepcopy(array)

print('Исходный массив:', interpolation_array)
print('Индекс запрошенного элемента:', interpolation_search(interpolation_array, 12))
print('Исходный массив:', interpolation_delete(interpolation_array, 50))
print('Индекс добавленного элемента:', interpolation_insert(interpolation_array, 66))
print('Исходный массив:', interpolation_array)

print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
```

### Простое рехэширование

```python
# O(1) - mid
# O(n) - worst

hash_name_list = [0 for i in range(len(name_list))]


def simple_hash(name, size, p = 0):
    unisum = 0
    for element in name:
        unisum += ord(element)      # Возвращает число из таблицы символов Unicode, представляющее его позицию
    return (unisum + p) % size


for name in name_list:
    i = simple_hash(name, len(name_list))
    p = 1
    while hash_name_list[i] != 0:
        i = simple_hash(name, len(name_list), p)
        p += 1
    hash_name_list[i] = name

print(hash_name_list, '\n')

# Вывод имени и его позиции в списке
for name in name_list:
    i = simple_hash(name, len(name_list))
    p = 1
    while hash_name_list[i] != name:
        i = simple_hash(name, len(name_list), p)
        p += 1
    print(hash_name_list[i], i)
```

### Рехэширование с помощью псевдослучайных чисел

```python
hash_name_list = [0 for i in range(len(name_list))]


def random_generate_num(name, size, p):
    a = 2
    b = 5
    return ((p ** b) // (size ** b)) + len(name) ** (b - a)


def random_hash(name, size, p = 0):
    unisum = 0
    for element in name:
        unisum += ord(element)      # Возвращает число из таблицы символов Unicode, представляющее его позицию
    return (unisum + random_generate_num(name, size, p)) % size


for name in name_list:
    i = random_hash(name, len(name_list))
    p = 1
    while hash_name_list[i] != 0:
        i = random_hash(name, len(name_list), p)
        p += 1
    hash_name_list[i] = name

print(hash_name_list, '\n')

# Вывод имени и его позиции в списке
for name in name_list:
    i = random_hash(name, len(name_list))
    p = 1
    while hash_name_list[i] != name:
        i = random_hash(name, len(name_list), p)
        p += 1
    print(hash_name_list[i], i)
```

### Метод цепочек

```python
class HashList():

    def __init__(self, data=None):
        self.data = data
        self.next = None

        
    def add_hashlist(self, data):
        if self.data is None:
            self.data = data
        elif self.next is None:
            self.next = HashList(data)
        else:
            self.next.add_hashlist(data)

            
    def print_hashlist(self):
        if self.data:
            print(self.data, end=' ')
            if self.next:
                self.next.print_hashlist()
            else:
                return


hash_name_list = [HashList() for i in range(len(name_list))]


def node_hash(name, size):
    unisum = 0
    for element in name:
        unisum += ord(element)      # Возвращает число из таблицы символов Unicode, представляющее его позицию
    return unisum % size


for name in name_list:
    i = node_hash(name, len(name_list))
    hash_name_list[i].add_hashlist(name)

# Вывод на экран списка
for i in range(len(hash_name_list)):
    print(i, end=' ')
    hash_name_list[i].print_hashlist()
    print()
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.


```python
board = [[0 for i in range(8)] for j in range(8)]
find_solutions = 0


def set_queen(i, j):
    for x in range(8):
        board[x][j] += 1                # Заполняем горизонталь
        board[i][x] += 1                # Заполняем вертикаль
        if 0 <= i + j - x < 8:          # Заполняем диагонали
            board[i + j - x][x] += 1
        if 0 <= i - j + x < 8:
            board[i - j + x][x] += 1
    board[i][j] = -1                    # Сам ферзь


def remove_queen(i, j):
    for x in range(8):
        board[x][j] -= 1
        board[i][x] -= 1
        if 0 <= i + j - x < 8:
            board[i + j - x][x] -= 1
        if 0 <= i - j + x < 8:
            board[i - j + x][x] -= 1
    board[i][j] = 0


def print_position():
    global find_solutions
    find_solutions += 1
    
    abc = 'ABCDEFGH'
    ans = []
    for i in range(8):
        for j in range(8):
            if board[i][j] == -1:
                ans.append(abc[j] + str(i + 1))
    print(', '.join(ans))


def solve(i):
    for j in range(8):
        if board[i][j] == 0:
            set_queen(i, j)
            if i == 7:
                print_position()
            else:
                solve(i + 1)
            remove_queen(i, j)
    

solve(0)

print('\nНайдено всего решений: ', find_solutions)
```
