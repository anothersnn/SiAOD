```python
#1

cnt = int(input())
a = list(range(cnt))

def foo(a):
    a = sorted(a, key=lambda x: -x)
    
    for i in range(len(a) - 2):
        if a[i] < a[i + 1] + a[i + 2]:
            return a[i], a[i + 1], a[i + 2]
        
    return 0

print(foo(a))
```

```python
#2

a = [1, 2, 3, 0, 9]

def foo(a):
    a = map(str, sorted(a, key=lambda x: -x))
    
    return ''.join(a)

print(foo(a))
```

```python
#3

mat = [[6, 3, 41, 72], [5, 87, 7, 98], [5, 10, 12, 0]]

def foo(mat):
    side = min(len(mat), len(mat[0])) 
    
    main_dg = sorted([mat[i][i] for i in range (side)])
    side_dg = sorted([mat[-i - 1][i] for i in range(side)])
    
    for i in range(len(main_dg)):
        mat[i][i] = main_dg[i]
        mat[-i - 1][i] = side_dg[i]
        
    return mat

print(*mat, sep='\n')
print()
print(*foo(mat), sep='\n')
```
